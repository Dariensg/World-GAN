```python
import seaborn as sns
import matplotlib.pyplot as plt
from utils import load_pkl
import pymde
import numpy as np
import torch

ruins_path = "/home/schubert/projects/World-GAN/input/minecraft/ruins/"

for embedding_dim in [8, 32]:
    for use_naive in [True, False]:
        fig = plt.figure(constrained_layout=True)
        name = f"natural_representations_small_{embedding_dim}{'_naive' if use_naive else ''}"
        reps = load_pkl(prepath=ruins_path, name=name)
        embedding = pymde.preserve_distances(torch.stack(list(reps.values())), embedding_dim=2).embed()
        ax = pymde.plot(embedding, marker_size=50, color_by=[t.replace("minecraft:", "").replace("_", " ") for t in reps.keys()])
        plt.savefig(f"{ruins_path}/{name}.png", bbox_inches = 'tight')
        plt.savefig(f"{ruins_path}/{name}.svg", bbox_inches = 'tight')
        plt.savefig(f"{ruins_path}/{name}.pdf", bbox_inches = 'tight')
```
